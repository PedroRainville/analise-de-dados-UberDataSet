#Tratamento de Transformação dos Dados

```python
df_uber["Date"] = pd.to_datetime(df_uber["Date"], dayfirst=True).dt.normalize()
df_uber["Time"] = pd.to_datetime(df_uber["Time"], format="%H:%M:%S").dt.time
```

```python
numericas = df_uber.select_dtypes(include=["int64", "float64"])
categoricas = df_uber.select_dtypes(include=["bool", "category", "object"])
```

#Separar valores para conseguir fazer a media para preencher os valores nulos

```python
for col in numericas.columns:
  df_uber[col] = df_uber[col].fillna(df_uber[col].median())
```

```python
for col in categoricas.columns:
  mode_values = df_uber[col].mode()[0]
  df_uber[col] = df_uber[col].fillna(mode_values)
```

#KPIS !!!



#1. Quais os horarios do dia com mais frequencia em corridas?

```python
df_uber["Time"] = pd.to_datetime(df_uber["Time"].astype(str)).dt.hour
```

```python
plt.figure()
sns.countplot(x="Time", data=df_uber)
plt.title("Quantidade de corridas por hora do dia")
plt.xlabel("Hora do dia")
plt.ylabel("Corridas")
plt.show()
```

Horário com maior volume de corridas ocorre às 18h, seguido por 17h e 19h, indicando que o período de maior demanda coincide com o horário de saída do trabalho.

#2. Dias da semana com maior volume de corridas

```python
df_uber["Date"] = pd.to_datetime(df_uber["Date"])

df_uber["DiaSemana"] = df_uber["Date"].dt.day_name()

traducao = {
    "Monday": "Segunda-feira",
    "Tuesday": "Terça-feira",
    "Wednesday": "Quarta-feira",
    "Thursday": "Quinta-feira",
    "Friday": "Sexta-feira",
    "Saturday": "Sábado",
    "Sunday": "Domingo"
}
df_uber["DiaSemana"] = df_uber["DiaSemana"].map(traducao)

dias_ordenados = ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira", "Sábado", "Domingo"]

contagem_dias = df_uber["DiaSemana"].value_counts().reindex(dias_ordenados)

cores = cm.viridis(np.linspace(0, 1, len(dias_ordenados)))

plt.figure(figsize=(8,8))
plt.pie(contagem_dias, labels=dias_ordenados, autopct="%1.1f%%", startangle=90, colors=cores)
plt.title("Distribuição de Corridas por Dia da Semana", fontsize=16)
plt.axis("equal")
plt.show()
```

#3. Dias da semana com mais cancelamentos

```python
df_uber["DiaSemana"] = df_uber["Date"].dt.day_name()
```

```python
plt.figure()
sns.countplot(x="DiaSemana", hue="Booking Status", data = df_uber)
plt.title("Quantidade de Cancelamentos por Dia da Semana")
plt.xlabel("Dia da Semana")
plt.ylabel("Cancelamentos")
plt.xticks(rotation=45)
plt.show()
```

#4. Motivo de cancelamento com base no Usuário e no Motorista

```python
total_trips = len(df_uber)

cancelled_by_customer = df_uber[df_uber["Booking Status"] == "Cancelled by Customer"].shape[0]
pct_cancelled_by_customer = cancelled_by_customer / total_trips * 100

cancelled_by_driver = df_uber[df_uber["Booking Status"] == "Cancelled by Driver"].shape[0]
pct_cancelled_by_driver = cancelled_by_driver / total_trips * 100

incomplete_trips = df_uber[df_uber["Booking Status"] == "Incomplete"].shape[0]

print(f"Percentual canceladas pelo cliente: {pct_cancelled_by_customer:.2f}%")
print(f"Percentual canceladas pelo motorista: {pct_cancelled_by_driver:.2f}%")
print(f"Viagens incompletas: {incomplete_trips}")

top_reasons_customer = (df_uber[df_uber["Cancelled Rides by Customer"] == 1]["Reason for cancelling by Customer"]
                        .value_counts().head(10))
top_reasons_driver = (df_uber[df_uber["Cancelled Rides by Driver"] == 1]["Driver Cancellation Reason"]
                      .value_counts().head(10))

plt.figure()
sns.barplot(x=top_reasons_customer.values, y=top_reasons_customer.index, palette="viridis")
plt.title("Motivos de Cancelamento pelo Cliente")
plt.xlabel("Número de cancelamentos")
plt.ylabel("Motivo")
plt.show()

print("Motivos de cancelamento (cliente):")
print(top_reasons_customer)

plt.figure()
sns.barplot(x=top_reasons_driver.values, y=top_reasons_driver.index, palette="magma")
plt.title("Motivos de Cancelamento pelo Motorista")
plt.xlabel("Número de cancelamentos")
plt.ylabel("Motivo")
plt.show()

print("Motivos de cancelamento (motorista):")
print(top_reasons_driver)

mean_vtat = df_uber["Avg VTAT"].mean()
mean_ctat = df_uber["Avg CTAT"].mean()
print(f"Média VTAT: {mean_vtat:.2f}")
print(f"Média CTAT: {mean_ctat:.2f}")
```

#5. Qual o metodo de pagamento mais usado e a receita anual?

```python
mean_value_per_trip = df_uber["Booking Value"].mean()
print(f"Valor médio por viagem: {mean_value_per_trip:.2f}")

start_date = "2024-01-01"
end_date = "2024-12-31"
mask = (df_uber["Date"] >= start_date) & (df_uber["Date"] <= end_date)
revenue_total = df_uber.loc[mask, "Booking Value"].sum()
print(f"Receita total entre {start_date} e {end_date}: {revenue_total:.2f}")

most_used_payment = df_uber["Payment Method"].value_counts().idxmax()
print(f"Método de pagamento mais usado: {most_used_payment}")

plt.figure()
sns.countplot(data=df_uber, x="Payment Method", order=df_uber["Payment Method"].value_counts().index, palette="pastel")
plt.title("Método de Pagamento mais utilizado")
plt.xticks(rotation=45)
plt.show()
```

#6. Qual foi a media de distância por corrida e os veiculos mais usados?

```python
mean_distance = df_uber["Ride Distance"].mean()
print(f"Distância média por corrida: {mean_distance:.2f}")

most_used_vehicle = df_uber["Vehicle Type"].value_counts().idxmax()
print(f"Tipo de veículo mais usado: {most_used_vehicle}")

plt.figure()
sns.countplot(data=df_uber, x="Vehicle Type", order=df_uber["Vehicle Type"].value_counts().index, palette="Set2")
plt.title("Tipo de Veículo mais usado")
plt.xticks(rotation=45)
plt.show()
```

#7. Quais foram os lugares com mais embarques?

```python
top_pickups = df_uber["Pickup Location"].value_counts().head(10)
print(top_pickups)

plt.figure()
sns.barplot(x = top_pickups.values, y = top_pickups.index, palette="coolwarm")
plt.title("Top 10 Locais de Pickup")
plt.xlabel("Número de viagens")
plt.ylabel("Local de Pickup")
plt.show()
```

#8. Taxa de Conclusão de Viagens (Trip Completion Rate)

```python
completed_trips = df_uber[df_uber["Booking Status"] == "Completed"].shape[0]
trip_completion_rate = completed_trips / total_trips * 100

print(f"Taxa de conclusão de viagens: {trip_completion_rate:.2f}%")

plt.figure()
sns.barplot(x=["Completas", "Não Completas"],
            y=[trip_completion_rate, 100 - trip_completion_rate],
            palette="crest")
plt.title("Taxa de Conclusão de Viagens (%)")
plt.ylabel("Percentual")
plt.show()
```

#9. Quais foram as rotas mais frequentes?

```python
df_uber["Route"] = df_uber["Pickup Location"] + " -> " + df_uber["Drop Location"]
top_routes = df_uber["Route"].value_counts().head(10)
sns.barplot(x=top_routes.values, y=top_routes.index, palette="viridis")
plt.title("Top 10 Rotas")
plt.show()

print("Top 10 rotas mais frequentes:")
print(top_routes)
```

#10. Qual foi o crescimento mensal de clientes no ano?

```python
df_uber["Month"] = df_uber["Date"].dt.to_period("M")
new_customers_per_month = df_uber.groupby("Month")["Customer ID"].nunique()
new_customers_per_month.plot(kind="line", title="Novos Clientes por Mês")
plt.ylabel("Número de Clientes")
plt.show()
```
